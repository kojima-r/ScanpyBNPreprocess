```python
import glob
WT=[]
AD=[]
for line in open("AD/info.tsv"):
    arr=line.strip().split("\t")
    if "WT6m_mouse" in arr[2]:
        target=arr[0]
        filenames=glob.glob("AD/data/*_"+target+".txt")
        filename = filenames[0]
        df=pd.read_csv(filename,sep="\t")
        df2=df.mean(axis=1)
        WT.append(df2)
    elif "AD6m_mouse" in arr[2]:
        target=arr[0]
        filenames=glob.glob("AD/data/*_"+target+".txt")
        filename = filenames[0]
        df=pd.read_csv(filename,sep="\t")
        df2=df.mean(axis=1)
        AD.append(df2)
    
```

```python
WT[1]
```

```python
WT_df=pd.concat(WT,axis=1)
AD_df=pd.concat(AD,axis=1)
```

```python
AD_df
```

```python
wt=WT_df.mean(axis=1)
ad=AD_df.mean(axis=1)

df_all=pd.concat([wt,ad],axis=1)
df_all.columns=["WT","AD"]
df_all
```

```python
gene_list=[]
for line in open("./03data_bbknn/all2.txt"):
    e=line.strip().split("\t")[0]
    gene_list.append(e)
    
```

```python
df_all_re=df_all.reindex(index=gene_list[1:])
df_all_re.index.name ="@name"
df_all_re=df_all_re.fillna(0)
df_all_re
```
